```python
# step 3 missing value ko fill krna 
df['total_bedrooms'].fillna(df['total_bedrooms'].median(), inplace=True)
```

```python
df['total_bedrooms'].isnull().sum()
```

```python
# ab hme jo apne data set me pridict krna hai wo nikalenge x and y ke through
y = df['median_house_value']
```

```python
y
```

```python
# isme jo column hm predict kr rhe us column ko drop krna baki sb ko chalaana
X = df.drop('median_house_value', axis=1)
```

```python
X
```

```python
# ab he encoading krni hai jis column me text value hai use categories krna aur o and 1 me todana 
x = pd.get_dummies(X, columns=['ocean_proximity'], drop_first=True)
print(x.shape)
```

```python
# ab hm data me train and test split krenge
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test, = train_test_split(x, y, test_size=0.2, random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
# ab hm scaling krne ke liye start kr rhe hai.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
print(X_train_scaled)
```

```python
# ab hm model banayenge 
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train_scaled, Y_train)
```

```python
# model train ho gya hai 
# ab hm predict krenge
y_pred = model.predict(X_test_scaled)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

rmse = np.sqrt(mean_squared_error(Y_test, y_pred))
r2 = r2_score(Y_test, y_pred)
```

```python
print("RMSE:", rmse)
print("R2 Score", r2)
```

```python
# ab hm random forest use krenge for better acccuracy model
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train_scaled, Y_train)
```

```python
y_pred_rf = rf_model.predict(X_test_scaled)
```

```python
rmse_rf = np.sqrt(mean_squared_error(Y_test,y_pred_rf))
r2_rf = r2_score(Y_test, y_pred_rf)
```

```python
print("RMS:", rmse_rf)
print("R2_score", r2_rf)
```
